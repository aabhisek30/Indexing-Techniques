```python
global overflow_data 
overflow_data = {}
```

```python
def new_hash_bucket(num):
    """
    It creats new bucket and handles data insertion
    """
    global split_pointer 
    new_hash_bucket_no = len(hashing_data)
    hashing_data[str(new_hash_bucket_no)] = []
    check_load_factor(num)
    
    split_data = hashing_data[str(split_pointer)]
    hashing_data[str(split_pointer)] = []
    
    for i in split_data:
        index = i % 8
        hashing_data[str(index)].append(i)
        
    if bool(overflow_data):
        for key,value in overflow_data.items():
            if str(key) == str(split_pointer):
                data = overflow_data[str(split_pointer)]
                for i in data:
                    index = i % 8
                    hashing_data[str(index)].append(i)
        del overflow_data[key]
    split_pointer +=1    
    return 
```

```python
def insert_in_bucket(num):
    """
    This insert value into bucket
    """
    hash_key = num % 4
    if len(hashing_data[str(hash_key)]) >= 2:
        if str(hash_key) not in overflow_data:
            overflow_data[str(hash_key)] = []
            overflow_data[str(hash_key)].append(num)
        else:
            overflow_data[str(hash_key)].append(num)
    else:
        hashing_data[str(hash_key)].append(num)
    return
```

```python
def check_load_factor(num):
    """
    It first count values in hashing data and overflow bucket. It than caclulate load factor. It make new bucket if load factor 
    is more than threeshold else insert in bucket
    """
    count = 0
    data_values = hashing_data.values()
    for values in data_values:
        for v1 in values:
            count = count + 1
    if bool(overflow_data):
        for k,v in overflow_data.items():
            for i in v:
                count += 1
    no_of_hash_bucket = len(hashing_data)
    if (load_factor<((count+1)/(2*no_of_hash_bucket))):
        new_hash_bucket(num)
    else:
        insert_in_bucket(num)
```

```python
def insert():
    """
    This function takes value to insert into hashing and calls check_load_factor function to check load factor
    """
    num = int(input("enter the number"))
    check_load_factor(num)
    print("Data after inserting")
    printf()
    input_type()
```

```python
def printf():
    """
    This function prints data in hashing and prints data of overflow bucket
    """
    for key,value in hashing_data.items():
        print(key,value)
        if key in overflow_data:
            print(key,overflow_data[key])
```

```python
def input_type():
    """
    This is the menu driver of the program
    """
    input_data = int(input("Enter 1 to insert,2 to print data,3 to quit"))
    if input_data == 1:
        insert()
    elif input_data == 2:
        printf()
    else:
        return 
```
